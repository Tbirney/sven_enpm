```python
# this script is used to search recursively in directories
# will do a depth first search, open files, and search for all 
```

```python
import os

data_dirs = [
    "target-pandas-src-code",
    "target-numpy-src-code",
    "pandas-test-code",
    "numpy-test-code"
]

all_functions = []

for src_dir in data_dirs:
    repo_functions = []

    for file in os.listdir(src_dir):

        with open(f"{src_dir}/{file}", 'r') as file:
            print(f"Reading {file} in {src_dir}")

            src_code = file.read()

            #split on function defininitions
            file_functions = src_code.split('def ')
            file_functions.pop(0)
            file_functions = ['def ' + x for x in file_functions]

            formatted_file_functions = []

            # filtering out other information that may not be part of the function
            for i, ff in enumerate(file_functions):
                split_by_line = ff.split('\n')
                split_by_line = list(filter(lambda x: (len(x) != len(x.strip()) or x.strip().startswith('def ')), split_by_line))

                formatted_file_functions.append('\n'.join(split_by_line))


            repo_functions += formatted_file_functions
            
    all_functions.append(repo_functions)
    print(len(repo_functions))

    with open(f"{src_dir}_functions.txt", 'w') as outfile:
        outfile.writelines(repo_functions)

```
